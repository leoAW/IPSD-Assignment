# 4. Drop Missing Value Berdasarkan Baris, Kolom, Imputasi dengan Mean, Modus, dan Median

A. Berdasarkan Baris

```python
baris = data.copy(deep=True)
drop_baris = baris.dropna(axis=0)
```

B. Berdasarkan Kolom

```python
kolom = data.copy(deep=True)
drop_kolom = kolom.dropna(axis=1)
```

C. Berdasarkan Imputasi dengan Mean

```python
mean_imputation = data.copy(deep=True)
mean_imputation['Time_taken'] = mean_imputation['Time_taken'].fillna(mean_imputation['Time_taken'].mean())
```

```python
mean_imputation.isna().sum()
```

D. Berdasarkan Imputasi dengan Modus

```python
modus_imputation = data.copy(deep=True)
modus_imputation['Time_taken'] = modus_imputation['Time_taken'].fillna(modus_imputation['Time_taken'].mode()[0])
```

```python
modus_imputation.isna().sum()
```

E. Berdasarkan Imputasi dengan Median

```python
median_imputation = data.copy(deep=True)
median_imputation['Time_taken'] = median_imputation['Time_taken'].fillna(median_imputation['Time_taken'].median())
```

```python
median_imputation.isnull().sum()
```
